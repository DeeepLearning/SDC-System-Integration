### Generating Proposed Clockwise waypoints using original bag loop vehicle positions as guide

```python
x_waypoint_p = x_waypoint_r[290:439] + x_waypoint_r[50:155] + x_waypoint_r[915:990]
y_waypoint_p = y_waypoint_r[290:439] + y_waypoint_r[50:155] + y_waypoint_r[915:990]


x_start_n = [ x_waypoint_p[0] ]
y_start_n = [ y_waypoint_p[0] ]
x_end_n = [ x_waypoint_p[-1]]
y_end_n = [ y_waypoint_p[-1]]

tl_x = [ 20.991 ]
tl_y = [ 22.837 ]
```

```python
plt.rcParams["figure.figsize"] = [16, 16]
p1 = plt.plot(x_waypoint_r, y_waypoint_r, 'go', ms=2)
p2 = plt.plot(x_waypoint, y_waypoint, 'b', lw=1.0)
p3 = plt.plot(x_waypoint, y_waypoint, 'go', ms=5.)
p4 = plt.plot(x_start, y_start, 'go', ms=10.0)
p5 = plt.plot(x_end, y_end, 'ro', ms=10.0)


p6 = plt.plot(x_start_n, y_start_n, 'go', ms=15.0)
p7 = plt.plot(x_end_n, y_end_n, 'ro', ms=15.0)
p8 = plt.plot(x_waypoint_p, y_waypoint_p, 'r', lw=2.0)
p9 = plt.plot(x_waypoint_p, y_waypoint_p, 'bo', ms=5)
p10 = plt.plot(tl_x, tl_y, 'ro', ms=20.0)


plt.title('SDC System Integration Carla Test Course Map from Udacity rosbag sample')
plt.xlabel("X", fontsize=10)
plt.ylabel("Y", fontsize=10)
plt.legend((p1[0], p2[0], p3[0], p4[0], p5[0], p6[0], p7[0], p8[0], p9[0], p10[0]), ('Recorded Rosbag Vehicle Positions', 'Previous Path', 'Previous Waypoints', 'Previous Starting Position', 'Previous Ending Position', 'Proposed New Starting Position', 'Proposed New Ending Position', 'Proposed New Path', 'Guiding Rosbag vehicle positions', 'Traffic Light'), loc=0)
plt.show()
```

```python
import math
def adistance(x, y, i):
    return math.sqrt((x[i]-x[i-1])**2 + (y[i]-y[i-1])**2)

x_waypoint_p = np.array(x_waypoint_p).astype(np.float32)
y_waypoint_p = np.array(y_waypoint_p).astype(np.float32)

x_waypoint_new = []
y_waypoint_new = []
s_waypoint_p = [ 0. ]
d = 0.

for i in range(len(x_waypoint_p)-1):
    d += adistance(x_waypoint_p, y_waypoint_p, i+1)
    s_waypoint_p.append(d)
s_waypoint_p = np.array(s_waypoint_p).astype(np.float32)

xpolyfit1 = np.polyfit(s_waypoint_p[:110], x_waypoint_p[:110], 2)
ypolyfit1 = np.polyfit(s_waypoint_p[:110], y_waypoint_p[:110], 2)
xpoly1 = np.poly1d(xpolyfit1)
ypoly1 = np.poly1d(ypolyfit1)

xpolyfit2 = np.polyfit(s_waypoint_p[100:], x_waypoint_p[100:], 16)
ypolyfit2 = np.polyfit(s_waypoint_p[100:], y_waypoint_p[100:], 12)
xpoly2 = np.poly1d(xpolyfit2)
ypoly2 = np.poly1d(ypolyfit2)

splice_point = 16

for i in range(splice_point):
    x_waypoint_new.append(xpoly1([float(i)])[0])
    y_waypoint_new.append(ypoly1([float(i)])[0])
    
for i in range(int(s_waypoint_p[-1]-splice_point)):
    x_waypoint_new.append(xpoly2([float(i+splice_point)])[0])
    y_waypoint_new.append(ypoly2([float(i+splice_point)])[0])
    
x_waypoint_new.append(x_end_n[0])
y_waypoint_new.append(y_end_n[0])

x_waypoint_mapped = []
y_waypoint_mapped = []
z_waypoint_mapped = []
yaw_waypoint_mapped = []

x_waypoint_r = np.array(x_waypoint_r).astype(np.float32)
y_waypoint_r = np.array(y_waypoint_r).astype(np.float32)

def nearest_waypoint_r(x, y):
    bd = 10000.
    bi = 0
    for i in range(len(x_waypoint_r)):
        d = math.sqrt((x_waypoint_r[i]-x)**2 + (y_waypoint_r[i]-y)**2)
        if d < bd:
            bd = d
            bi = i
    return bi

for i in range(len(x_waypoint_new)):
    x = float(x_waypoint_new[i])
    y = float(y_waypoint_new[i])
    p = nearest_waypoint_r(x, y)
    x_waypoint_mapped.append(x_waypoint_r[p])
    y_waypoint_mapped.append(y_waypoint_r[p])
    z_waypoint_mapped.append(z_waypoint_r[p])
    yaw_waypoint_mapped.append(yaw_waypoint_r[p])
    
ftl_x = [ x_waypoint_new[16] ]
ftl_y = [ y_waypoint_new[16] ]
```

```python
plt.rcParams["figure.figsize"] = [16, 16]
p1 = plt.plot(x_waypoint_r, y_waypoint_r, 'go', ms=2)
p2 = plt.plot(x_waypoint, y_waypoint, 'b', lw=1.0)
p3 = plt.plot(x_waypoint, y_waypoint, 'go', ms=5.)
p4 = plt.plot(x_start, y_start, 'go', ms=10.0)
p5 = plt.plot(x_end, y_end, 'ro', ms=10.0)

p6 = plt.plot(x_start_n, y_start_n, 'go', ms=15.0)
p7 = plt.plot(x_end_n, y_end_n, 'ro', ms=15.0)
p8 = plt.plot(x_waypoint_new, y_waypoint_new, 'r', lw=1.0)
p9 = plt.plot(x_waypoint_new, y_waypoint_new, 'bo', ms=5)
p10 = plt.plot(tl_x, tl_y, 'ro', ms=20.0)


plt.title('SDC System Integration Carla Test Course Map from Udacity rosbag sample')
plt.xlabel("X", fontsize=10)
plt.ylabel("Y", fontsize=10)
plt.legend((p1[0], p2[0], p3[0], p4[0], p5[0], p6[0], p7[0], p8[0], p9[0], p10[0]), ('Recorded Rosbag Vehicle Positions', 'Previous Path', 'Previous Waypoints', 'Previous Starting Position', 'Previous Ending Position', 'Proposed New Starting Position', 'Proposed New Ending Position', 'Proposed New Path', 'Proposed New Waypoint Positions', 'Traffic Light'), loc=0)
plt.show()
```
