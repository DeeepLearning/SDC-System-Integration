```python
import os
from keras.preprocessing.image import load_img,img_to_array
import numpy as np
from train import SqueezeNet
from consts import IMAGE_WIDTH, IMAGE_HEIGHT
```

```python
# load pretrained model
model = SqueezeNet(3, (IMAGE_HEIGHT, IMAGE_WIDTH, 3))
model.load_weights("challenge1.weights")
```

```python
# predict image
#filepath = "../02-images_from_simulator/sampleout1.jpg"
filepath = "../03-Udacity-trafficlights/just_traffic_light_0180.jpg"
#TEST_FOLDER="/Users/rainerbareiss/Library/Mobile Documents/M6HJR9W95L~com~textasticapp~textastic/Documents/02-Udacity/carND/03-term3/P3-CarND-Capstone/00-team_vulture/SDC-System-Integration/classifier/03-Udacity-trafficlights"

image = load_img(filepath, target_size=(224,224))
image = img_to_array(image)
image /= 255.0
image = np.expand_dims(image, axis=0)
preds = model.predict(image)[0]
predclass = str(np.argmax(preds))
```

```python
print(predclass)
# predclass
# 0: no traffic light
# 1: red light
# 2: green light
```
